```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\cust_satisfication.csv\cust_satisfaction.csv")
```

```python
print(df.head())
```

check null values

```python
print("\nNull Values:\n")
print(df.isnull().sum())
```

Remove null values

```python
df = df.dropna()
```

Remove duplicate values

```python
df = df.drop_duplicates()
print("After Cleaning Shape:", df.shape)
```

Filter data

```python
filtered_data = df[df["Age"] > 30]

print(filtered_data.head())
```

```python
filtered_data = df[df["Flight Distance"] > 1000]

print(filtered_data.head())
```

Save cleaned data

```python
filtered_data.to_csv("cleaned_cust_satisfaction.csv", index=False)
```

create graphs

```python
df["satisfaction"].value_counts().plot(kind="bar")

plt.title("Customer Satisfaction Count")
plt.xlabel("Satisfaction")
plt.ylabel("Count")
plt.show()
```

```python
df["Age"].plot(kind="hist", bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["Age"], df["Flight Distance"])

plt.title("Age vs Flight Distance")
plt.xlabel("Age")
plt.ylabel("Flight Distance")
plt.show()
```
